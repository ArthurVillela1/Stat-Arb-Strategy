```python
import random
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import coint, adfuller
from statsmodels.regression.linear_model import OLS
```

```python
# Fetch Brazilian stock tickers from a reliable source
def get_brazilian_tickers():
    url = 'https://en.wikipedia.org/wiki/List_of_companies_listed_on_B3'
    tables = pd.read_html(url)
    # Manually set the column names
    columns = ['Código', 'Ação', 'Tipo', 'Qtde Teorica', 'Part. (%)']
    tables[0].columns = columns
    # Use the correct column name 'Código'
    tickers = tables[0]['Código'].tolist()
    tickers = [ticker + '.SA' for ticker in tickers]
    return tickers
```

```python
# Fetch historical data for Brazilian stocks
def get_historical_data(tickers):
    data = pd.DataFrame()
    for ticker in tickers:
        try:
            stock_data = yf.download(ticker, start="2022-09-30", end="2024-09-30")
            if 'Adj Close' in stock_data.columns:
                data[ticker] = stock_data['Adj Close']
            else:
                print(f"Failed to download data for {ticker}")
        except Exception as e:
            print(f"Error downloading data for {ticker}: {e}")
    return data
```

```python
# Example usage
brazilian_tickers = get_brazilian_tickers()
historical_data = get_historical_data(brazilian_tickers)  # Fetch data for all tickers

# Set pandas display options
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

# Display the entire DataFrame
print(brazilian_tickers)
display(historical_data)
```
